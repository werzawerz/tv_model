```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df_read = pd.read_parquet(f"tv_data.parquet")
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 50)
df_read
```

```python
df_read.describe()
```

```python
# Remove ":" from column_names
df_read.columns = df_read.columns.str.replace(":", "")
```

The following changes need to be made to the features:
* Where we have unit of measure present in the field, remove it, so we can create numerical features
* Feature where "Igen" is a frequent value, convert Igen to 1, and null to 0
* Remove features where the value is empty in more than 50% of the cases
* Remove features where there is only one unique value
* Target -> Price
* ID -> Cikkszám

```python
# Remove features where the value is empty in more than 50% of the vases
df = df_read.dropna(axis=1, thresh=0.5 * len(df_read))
```

```python
# Remove features where all values are same, and there are no missing values
cols_to_remove = [col for col in df.columns if df[col].nunique() == 1 and df[col].count() == len(df)]
df = df.drop(columns = cols_to_remove)
```

```python
df
```

```python
features_with_unit_of_measure = [
    "Képernyőátló",
    "Képernyőátló (col)",
    "Zenei teljesítmény",
    "Szélesség állvány nélkül",
    "Magasság állvány nélkül",
    "Mélység állvány nélkül",
    "Szélesség állvánnyal",
    "Magasság állvánnyal",
    "Mélység állvánnyal",
    "Súly állvány nélkül",
    "Súly állvánnyal",
    "On mode power demand (SDR)",
    "Energiafogyasztás standard dinamikatartományban 1000 óránként (SDR)",
    "On mode power demand (HDR)",
    "Energiafogyasztás magas dinamikatartományban 1000 óránként (HDR)",
    "Képfrissítés"
]
```

```python
# Define a function to remove units of measure
def remove_unit(value):
    if value is None:
        return None
    return ''.join(char for char in value if not char.isalpha()).strip()

for col in features_with_unit_of_measure:
    print(col)
    df[col] = df[col].apply(remove_unit)
```

```python
df
```

```python
# Fill full of "Igen" values
for col in df.columns:
    if df[col].nunique() == 1 and "Igen" in df[col].unique():
        df[col].fillna("Nem", inplace=True)
```

```python
df.describe()
```

```python
features = [f for f in df.columns if f not in ["price", "Cikkszám"]]
```

```python
# Bool features to int
for col in df.columns:
    if df[col].nunique() == 2 and "Igen" in df[col].unique():
        df[col] = np.where(df[col] == "Igen", 1, 0)
```

```python
df
```

```python
# Convert column to numeric where possible
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
categorical_features = [f for f in features if df[f].dtype == "object"]
```

```python
len(categorical_features)
```

```python
numerical_features = [f for f in features if f not in categorical_features]
```

```python
df.dtypes
```

```python
# Fill categorical variables with a new category and numerical with median
for col in df.columns:
    if col in categorical_features:
        df[col].fillna("Dummy category", inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)
```

```python
df
```

```python
df.to_parquet("tv_data_processed.parquet")
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[features], df["price"], test_size=0.2, random_state=42)
```

## Catboost
Half of the features are categorical, it's convinient to use an algorithm which handles categorical variables 

```python
from catboost import CatBoostRegressor
```

```python
params = {
    "iterations": 10,
    "verbose": True,
    "depth": 3,
    "one_hot_max_size": 2,
    "random_seed": 42,
    "cat_features": categorical_features,
}
```

```python
model = CatBoostRegressor(**params)
```

```python
model.fit(X_train, y_train)
```

```python
df_imp = pd.DataFrame(
    np.array([features, model.feature_importances_]).T,
    columns=["feature", "impact"],
)
df_imp["impact"] = df_imp["impact"].astype(float)
df_imp.sort_values("impact", ascending=False, inplace=True)
```

```python
df_imp.head(5)
```

```python
y_pred_train = model.predict(X_train)
```

```python
rmse_train = mean_squared_error(y_train, y_pred_train, squared=False)
print(rmse_train)
```

```python
y_pred_test = model.predict(X_test)
```

```python
y_pred_test = model.predict(X_test)
rmse_test = mean_squared_error(y_test, y_pred_test, squared=False)
print(rmse_test)
```

## Try with a simpler model - logistic regression

```python
df_for_cat = df.copy()
```

```python
for cat_feature in categorical_features:
    # Calculate mean target values for each category in the training set
    mean_target = df_for_cat.groupby(cat_feature)['price'].mean()
    
    # Apply the encoding to both the training and validation/test sets
    df_for_cat[cat_feature] = df_for_cat[cat_feature].map(mean_target)
```

```python
df_for_cat
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_for_cat[features], df_for_cat["price"], test_size=0.2, random_state=42)
```

```python
X_train=(X_train-X_train.mean())/X_train.std()
X_test=(X_test-X_test.mean())/X_test.std()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
y_pred_train = clf.predict(X_train)
rmse_train = mean_squared_error(y_train, y_pred_train, squared=False)
print(rmse_train)
```

```python
X_test.fillna(0, inplace=True)
```

```python
y_pred_test = clf.predict(X_test)
rmse_test = mean_squared_error(y_test, y_pred_test, squared=False)
print(rmse_test)
```

Linear regression was better on the train dataset, but overfitted badly, the RMSE on the test dataset was lhigher, then during the catboost case
